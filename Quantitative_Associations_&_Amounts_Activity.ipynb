```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

#**Correlation Matrix**

Create a heatmap that displays the correlation between each of the quantitative variables in df using Altair.

```python
df = pd.read_csv("https://raw.githubusercontent.com/ef2020/TextFiles/master/iris.csv")
df.head()
```

```python
correlation_matrix=df.iloc[:, :len(df.columns)-1].corr()
correlation_matrix_with_index = correlation_matrix.reset_index()
correlation_matrix_with_index
```

```python
corr_melt = correlation_matrix_with_index.melt(id_vars='index', var_name='column', value_name='correlation')
```

```python
#TODO
heatmap = alt.Chart(corr_melt).mark_rect().encode(
    alt.X('column:N', sort='descending'),
    alt.Y('index:N', sort='descending'),
    color=alt.Color('correlation') 
).properties(width = 500, height = 500)
```

#**Stacked vs. Grouped Bar Chart**

```python
tips = sns.load_dataset("tips")
tips.head()
```

Create a stacked bar chart displaying the total_bill amount for smokers versus non-smokers by day.

```python
#TODO
stacked = alt.Chart(tips).mark_bar().encode(
    x = alt.X('day', axis=alt.Axis(labelAngle=0),  sort=['Thur', 'Fri', 'Sat', 'Sun']),
    y = 'sum(total_bill):Q',
    color='smoker:N',
).interactive().properties(width = 500)
stacked
```

Create a grouped bar chart displaying the total_bill amount for smokers versus non-smokers by day.

```python
#TODO
grouped = alt.Chart(tips).mark_bar().encode(
    x = alt.X('day', axis=alt.Axis(labelAngle=0), sort=['Thur', 'Fri', 'Sat', 'Sun']),
    y = 'sum(total_bill):Q',
    color='smoker:N',
    xOffset='smoker'
).interactive().properties(width= 500, title= "Smoker vs. Non-Smoker Total Bill Amount by Day of Week")
grouped
```

```python
grouped2 = alt.Chart(tips).mark_bar().encode(
    x = alt.X('smoker', axis=alt.Axis(labelAngle=0), sort=['No', 'Yes']),
    y = 'sum(total_bill):Q',
    color='day',
    xOffset='day'
).interactive().properties(width= 500, title= "Smoker vs. Non-Smoker Total Bill Amount by Day of Week")
grouped2
```
